# The Set up

```python
import requests
from bs4 import BeautifulSoup
```

```python
base_site = 'https://www.imdb.com/chart/top/?ref_=nv_mv_250'
```

```python
response = requests.get(base_site)
response
```

```python
html = response.content
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
with open('IMDb_top_rated_movies.html', 'wb') as file:
    file.write(soup.prettify('utf-8'))
```

# Obtaining the element containing all the data

```python
tables = soup.find('table', {'class' : 'chart full-width', 'data-caller-name': 'chart-top250movie'})
tables
```

# Extracting the title, year and rating of each movie

# Extracting the title

```python
titles = tables.find_all('td', {'class' : 'titleColumn'})
titles
```

```python
movie_names = [title.find('a').string for title in titles]
movie_names
```

# Year

```python
movie_years = [title.find('span').string for title in titles]
movie_years
```

# Removing brackets from year

```python
movie_years = [year.strip('()') for year in movie_years]
movie_years
```

```python
movie_years = [int(year) for year in movie_years]
movie_years
```


# Ratings

```python
ratings = tables.find_all('td', {'class' : 'ratingColumn imdbRating'})
ratings
```

```python
movie_rating = [rating.find('strong').string for rating in ratings]
movie_rating
```

```python
movie_rating = [float(rating) for rating in movie_rating]
movie_rating
```

# Represnting the data in structured form

```python
import pandas as pd
```

# Creating a DataFrame and populating it

```python
movies_info = pd.DataFrame()

movies_info['Movie Title'] = movie_names
movies_info['Year'] = movie_years
movies_info['Ratings'] = movie_rating

movies_info
```
